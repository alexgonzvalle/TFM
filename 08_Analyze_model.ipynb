```python
df = pd.read_csv('out_model_kmeans.csv')

# Ordenar por 'loc', 'model' y 'mse'
df = df.sort_values(by=['loc', 'model', 'mse'], ascending=[True, True, True])

# Crear la columna de ranking
df['rank'] = df.groupby(['loc', 'model'])['mse'].rank(method='first')

# Asegurarte de que los rankings sean enteros
df['rank'] = df['rank'].astype(int)

df_grouped = df.groupby(['k']).agg(
    ranking=('rank', 'sum')
).reset_index()
df_grouped.sort_values(by=['ranking'], ascending=[True])
```
